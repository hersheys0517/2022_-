```python
# 텍스트 파일 28부터 35까지 불러오기
lines2 = []
for i in range(28, 36):
    if i==54:
        i = i+1
    f = open("../피싱 사례 텍스트 파일/case0"+str(i)+".txt", 'r')
    lines = f.readlines()
    lines2 = lines2 + lines
    f.close()
```
