```python
import pandas as pd
import re
import csv
```

```python
# 텍스트 파일 28부터 100까지 불러오기
lines2 = []
for i in range(28, 100):
    if i==54:
        i = i+1
    f = open("../피싱 사례 텍스트 파일/case0"+str(i)+".txt", 'r')
    lines = f.readlines()
    lines2 = lines2 + lines # 28~100번 까지의 텍스트 파일
    f.close()
```

```python
# 텍스트 파일 101부터 142까지 불러오기
for i in range(100, 143):
    if i==137:
        i = i+1
    f = open("../피싱 사례 텍스트 파일/case"+str(i)+".txt", 'r')
    lines = f.readlines()
    lines2 = lines2 + lines
    f.close()
```

```python
# 텍스트 파일의 전체 list 원소들 중에 '\n'이 있을 시, 제거!
i=0
while i<len(lines2):
    if lines2[i]=='\n':
        del lines2[i]
        i = i-1
    i = i+1
```

```python
# 문장 뒤 개행 문자 제거
str_list = []
for i in lines2:
    i = i.strip()
    str_list.append(i)
str_list
```

```python
# 결과를 csv 파일로 저장
df = pd.DataFrame.from_dict(str_list)
df.to_csv('word_network_csv.csv')
```
